## One sample t test
## Exercise 1 : A company claims that on an average it takes only 40 hours to
## process any purchase order. Based on the data given below, can you validate
## the claim? The data is given in PO_Processing.csv

## H0 = mean is equal 40 hours
## H1 = mean is not equal to 40 hours

```python
# default is 0.05 significance
tscore , pvalue = ttest_1samp(df['Processing_Time'], popmean=40)
print("t Statistic: ", tscore)  
print("P Value:", pvalue)
alpha = 0.05
if (pvalue > alpha):
    print("(mean equals to 40) Failed to reject or do not reject null hypothesis")
else:
    print("(mean not equals to 40) Reject null hypothesis")
```

```python
# Perform one sample t-test using bioinfokit python
from bioinfokit.analys import stat
res = stat()
res.ttest(df=df, test_type=1, res='Processing_Time', mu=40, alpha=0.05)
print(res.summary)
```

```python
# Using pingouin
from pingouin import ttest
ttest(df['Processing_Time'], 40).round(4)
```

##  A computer manufacturing company claims that on an average it
## will respond to any complaint logged by the customer from anywhere in the world
## in 24 hours. Based on the data, validate the claim?

```python
df = pd.read_csv('.\\Data\\Complaint_Response_Time.csv')
df
```

```python
## H0 = mean is equal 24 hours
## H1 = mean is not equal to 24 hours
```

```python
# default is 0.05 significance
tscore , pvalue = ttest_1samp(df['Response_Time'], popmean=24)
print("t Statistic: ", tscore)  
print("P Value:", pvalue)
alpha = 0.05
if (pvalue > alpha):
    print("(mean equals to 24) Failed to reject or do not reject null hypothesis")
else:
    print("(mean not equals to 24) Reject null hypothesis")
```

```python
# Perform one sample t-test using bioinfokit python
from bioinfokit.analys import stat
res = stat()
res.ttest(df=df, test_type=1, res='Response_Time', mu=24, alpha=0.05)
print(res.summary)
```

```python
# Using pingouin
from pingouin import ttest
ttest(df['Response_Time'], 24).round(4)
```

## Paired or dependent t test

## The manager of a fleet of automobiles is testing two brands of radial tires. He
## assigns one tire of each brand at random to the two rear wheels of eight cars
## and runs the cars until the tire wear out. Is both brands have equal mean life?
## The data in kilometers is given in paired_ttest.csv

```python
df = pd.read_csv('.\\Data\\paired_ttest.csv')
df
```

```python
ttest_rel(df['Brand 1'], df['Brand 2'])
```

```python
# Perform two sample unpaired or independent t-test using bioinfokit python
from bioinfokit.analys import stat
res = stat()
res.ttest(df=df, res=['Brand 1','Brand 2'], test_type=3, alpha=0.05)
print(res.summary)
```

```python
# Using pingouin
from pingouin import ttest
ttest(df['Brand 1'], df['Brand 2'], paired=True).round(4)
```
